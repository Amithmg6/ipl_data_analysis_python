Premium by age

```python
ipl_data_auc_pre_age=ipl_data_auc.groupby('AGE')['premium'].mean().reset_index()
```

```python
print(ipl_data_auc_pre_age)
```

```python
print(ipl_data_auc.columns)
```

```python
ipl_data_auc_pre_age_country=ipl_data_auc.groupby('COUNTRY')['SOLD PRICE'].mean().round().sort_values(ascending=False).reset_index()
```

```python
ipl_data_auc_pre_playrole_country=ipl_data_auc.groupby(['PLAYING ROLE','COUNTRY'])['SOLD PRICE'].median().round(decimals=2).sort_values(ascending=False).reset_index()
```

joining the premium by country for age and playing roles

```python
ipl_data_auc_comparision=ipl_data_auc_pre_age_country.merge(ipl_data_auc_pre_playrole_country,on='COUNTRY',how='outer')
```

```python
ipl_data_auc_comparision
```

```python
ipl_data_auc_comparision.columns#rename(columns={''})
```

```python
ipl_data_auc_comparision.rename(columns={'SOLD PRICE_x':'SOLD_PRICE_age_country','SOLD PRICE_y':'SOLD_PRICE_playing_country'},inplace=True)
```

```python
ipl_data_auc_comparision['CHANGE']=ipl_data_auc_comparision.apply(lambda rec:(rec.SOLD_PRICE_age_country-rec.SOLD_PRICE_playing_country) / rec.SOLD_PRICE_playing_country, axis=1)
```

```python
ipl_data_auc_comparision.columns
```

```python
ipl_data_auc_comparision.sort_values('CHANGE',ascending=False).head(5)
```
