The price column automatically averages the data but we can do a count or a sum. Adding them is simple using aggfunc and np.sum

```python
pd.pivot_table(df,index=['PLAYING ROLE','AGE'],values=['BASE PRICE'],aggfunc=np.median)
```

```python
pd.pivot_table(df,index=['PLAYING ROLE','AGE'],values=['BASE PRICE'],aggfunc=[np.median,len])
```

```python
pd.pivot_table(df,index=['PLAYING ROLE','AGE'],values=['BASE PRICE'],
               columns=['COUNTRY'],aggfunc=np.median)
```

The NaN’s are a bit distracting. If we want to remove them, we could use fill_value to set them to 0.

```python
pd.pivot_table(df,index=['PLAYING ROLE','AGE'],
              values=['BASE PRICE'],columns=['COUNTRY'], aggfunc=np.median,fill_value=0)
```

```python
pd.pivot_table(df, index=['PLAYING ROLE','AGE'],values=['BASE PRICE'],columns=['COUNTRY','AUCTION YEAR'],aggfunc=np.median,fill_value=0)
```

```python
table=pd.pivot_table(df, index=['PLAYING ROLE','AGE'],
               values=['BASE PRICE'],
               columns=['COUNTRY','AUCTION YEAR'],
               aggfunc={'BASE PRICE':np.mean},
               fill_value=0,
               margins=True)

table
```

Enough fun with the tables, lets get going with new features which can be extracted from the data.

```python
df['PREMIUM']=df['SOLD PRICE']-df['BASE PRICE']
```

```python
soldprice_by_age=df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
soldprice_by_age.head()
```

```python
soldprice_by_age_role=df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
soldprice_by_age_role.head()
```

Joining column

```python
sp_comp=soldprice_by_age_role.merge(soldprice_by_age,
                                   on='AGE',
                                   how='outer')
sp_comp
```

```python
sp_comp.rename(columns={'SOLD PRICE_x':'Sold_price_age_role','SOLD PRICE_y':'Sold_price_age'},inplace=True)
```

```python
sp_comp.columns
```

```python
sp_comp['Change']=sp_comp.apply(lambda x:(x.Sold_price_age_role-x.Sold_price_age)/x.Sold_price_age,axis=1)
```
